```python
#import opencv
# To performing path manipulations 
import os
# Local Binary Pattern function
from skimage.feature import local_binary_pattern
# To calculate a normalized histogram 
from scipy.stats import itemfreq

from sklearn.preprocessing import normalize
# Utility package -- use pip install cvutils to install
import cvutils
# To read class from file
import csv
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
train_dic = {}
import pandas as pd
dataset = pd.read_csv('sample_labels2.csv',header = None)
for i in range (0,len(dataset[3])):
    train_dic["images/"+dataset[0][i]] = dataset[4][i]
```

```python
train_dic
```

```python
train_images = "images/"+dataset[0]
```

```python
import cv2
```

```python
X_test = []
X_name = []
y_test = []

from skimage.io import imread
# For each image in the training set calculate the LBP histogram
# and update X_test, X_name and y_test
for train_image in train_images:
    # Read the image
    #im = opencv.imread(train_image)
    im = imread(train_image)
    # Convert to grayscale as LBP works on grayscale image
    if(im.shape == (1024,1024)):
        im_gray = im
    else:
        im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    # Initiate STAR detector
    orb = cv2.ORB_create()
    # find the keypoints with ORB
    kp = orb.detect(im_gray,None)
    # compute the descriptors with ORB
    kp, x = orb.compute(im_gray, kp)#here x is the descriptor
    x2 = x.ravel()
    # Normalize the histogram
    x3 = x[:, 1]/sum(x[:, 1])
    # Append image path in X_name
    X_name.append(train_image)
    # Append histogram to X_name
    X_test.append(x3)
    # Append class label in y_test
    y_test.append(train_dic[train_image])

```

```python
import pandas as pd 
df = pd.DataFrame(X_test)
df.to_csv("X_orb_desc.csv")
```
